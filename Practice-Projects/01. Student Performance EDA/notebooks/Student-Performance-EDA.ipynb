```python
# using a custom style for my visualizations

plt.style.use('..\\references\\mystyle-v1.mplstyle')
fig_size = (18, 6)
```

```python
# getting all numerical variables
numerical_columns = []
for col in df.columns:
  if df[col].dtype != 'O':
    numerical_columns.append(col)
print(numerical_columns)
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2, )
plt.suptitle('Distributions of Numerical variables', )

columns = numerical_columns

for col, ax in zip(columns, axes.flatten()):
    sns.histplot(data=df, x=col, bins=20, kde=True, ax=ax)
    ax.set_title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2)
plt.suptitle('Boxplots Numerical variables')

columns = numerical_columns
for col, ax in zip(columns, axes.flatten()):
    ax.grid()
    ax.boxplot(df[col],vert = False, showmeans= True)
    ax.set_title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python

fig, axes = plt.subplots(nrows=4, ncols=2)
plt.suptitle('Distributions of Numerical Variables', y=1.05)

columns = numerical_columns

# Loop through each column and create subplots
for col, ax in zip(columns, axes.flatten()):
    # Create a new inset axes for the box plot
    ax_hist = ax  # Main axes for histogram
    ax_box = ax.inset_axes([0, 1.05, 1, 0.25])  # Inset for the box plot, positioned above the histogram

    # Plot histogram with KDE
    sns.histplot(data=df_mod1, x=col, bins=20, kde=True, ax=ax_hist)
    ax_hist.set_title(f'Distribution of {col}')

    # Plot box plot on top of the histogram
    ax_box.boxplot(df_mod1[col], vert=False, showmeans=True)
    ax_box.set_xticks([])  # Remove x-ticks for the box plot to avoid clutter
    ax_box.set_yticks([])  # Remove y-ticks as well

    # Adjust spacing for a cleaner layout
    plt.subplots_adjust(hspace=0.5)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=7, ncols=1,figsize= (12,25))
plt.suptitle('Bar Plots of Numerical variables', )

columns = numerical_columns

for col, ax in zip(columns, axes.flatten()):
    counts = df_mod1[col].value_counts().sort_index()
    counts.plot(kind='bar', ax=ax)
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# getting the catagorical variables
catagorical_columns = []
for col in df.columns:
  if df[col].dtype == 'O':
    catagorical_columns.append(col)
print(catagorical_columns)
```

```python
# total number of rows in dataset
num_rows = df_mod1.shape[0]
num_rows
```

```python
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(12, 12))
plt.suptitle('Bar Plots of Categorical Variables')

columns = catagorical_columns
num_rows = num_rows

for col, ax in zip(columns, axes.flatten()):
    col_value_counts = df_mod1[col].value_counts()
    col_value_groups = col_value_counts.index
    percentages = (col_value_counts / num_rows) * 100
    
    sns.barplot(x=col_value_groups, y=percentages, ax=ax)
    ax.set_xlabel(col)
    ax.set_ylabel('Percentage')

plt.tight_layout()  
plt.show()
```

Correlation analysis

```python
fig, axes = plt.subplots(nrows=4, ncols=2, figsize = (12,12))
plt.suptitle('correlation of Numerical variables with total')

columns = numerical_columns
target_var = 'Exam_Score'

for col, ax in zip(columns, axes.flatten()):
    ax.scatter(df[target_var],df[col])
    ax.set_title(f'correlation btween {col} and {target_var}')
    ax.set_xlabel(target_var)
    ax.set_ylabel(col)
    correlation = df[target_var].corr(df[col])
    corr_val = f'r = {correlation:.2f}'
    ax.text(df[target_var].min(), df[col].max() * 0.9, corr_val, fontsize=12, color='black', bbox=dict(facecolor='#d9d9d9', edgecolor='#636363', boxstyle='round,pad=0.5'))
  
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(10, 8))
plt.suptitle('correlation of Numerical variables with total')

columns = numerical_columns
correlation_matrix = df[columns].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='Oranges', fmt='.2f', cbar=True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
