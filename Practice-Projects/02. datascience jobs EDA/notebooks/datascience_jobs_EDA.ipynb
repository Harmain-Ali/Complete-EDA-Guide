‚öì <b>About the Dataset</b></br>
<div>
<p><strong>work_year</strong>: The year in which the data was recorded. This field indicates the temporal context of the data, important for understanding salary trends over time.</p>
<p><strong>job_title</strong>: The specific title of the job role, like 'Data Scientist', 'Data Engineer', or 'Data Analyst'. This column is crucial for understanding the salary distribution across various specialized roles within the data field.</p>
<p><strong>job_category</strong>: A classification of the job role into broader categories for easier analysis. This might include areas like 'Data Analysis', 'Machine Learning', 'Data Engineering', etc.</p>
<p><strong>salary_currency</strong>: The currency in which the salary is paid, such as USD, EUR, etc. This is important for currency conversion and understanding the actual value of the salary in a global context.</p>
<p><strong>salary</strong>: The annual gross salary of the role in the local currency. This raw salary figure is key for direct regional salary comparisons.</p>
<p><strong>salary_in_usd</strong>: The annual gross salary converted to United States Dollars (USD). This uniform currency conversion aids in global salary comparisons and analyses.</p>
<p><strong>employee_residence</strong>: The country of residence of the employee. This data point can be used to explore geographical salary differences and cost-of-living variations.</p>
<p><strong>experience_level</strong>: Classifies the professional experience level of the employee. Common categories might include 'Entry-level', 'Mid-level', 'Senior', and 'Executive', providing insight into how experience influences salary in data-related roles.</p>
<p><strong>employment_type</strong>: Specifies the type of employment, such as 'Full-time', 'Part-time', 'Contract', etc. This helps in analyzing how different employment arrangements affect salary structures.</p>
<p><strong>work_setting</strong>: The work setting or environment, like 'Remote', 'In-person', or 'Hybrid'. This column reflects the impact of work settings on salary levels in the data industry.</p>
<p><strong>company_location</strong>: The country where the company is located. It helps in analyzing how the location of the company affects salary structures.</p>
<p><strong>company_size</strong>: The size of the employer company, often categorized into small (S), medium (M), and large (L) sizes. This allows for analysis of how company size influences salary.</p></div>
</div>
