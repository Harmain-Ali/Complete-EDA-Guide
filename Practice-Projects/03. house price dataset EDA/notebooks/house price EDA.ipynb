<p><strong>Square_Footage</strong>: The size of the house in square feet. Larger homes typically have higher prices.</p>
<p><strong>Num_Bedrooms</strong>: The number of bedrooms in the house. More bedrooms generally increase the value of a home.</p>
<p><strong>Num_Bathrooms</strong>: The number of bathrooms in the house. Houses with more bathrooms are typically priced higher.</p>
<p><strong>Year_Built</strong>:The year the house was built. Older houses may be priced lower due to wear and tear.</p>
<p><strong>Lot_Size</strong>: The size of the lot the house is built on, measured in acres. Larger lots tend to add value to a property.</p>
<p><strong>Garage_Size</strong>: The number of cars that can fit in the garage. Houses with larger garages are usually more expensive.</p>
<p><strong>House_Price</strong>: The actual price of the House.</p></div>
