```python
# using a custom style for my visualizations

plt.style.use('..\\references\\mystyle-v1.mplstyle')
fig_size = (18, 6)
```

```python
# getting all numerical variables
numerical_columns = []
for col in df.columns:
  if df[col].dtype != 'O':
    numerical_columns.append(col)
print(numerical_columns)
```

```python

# List of numerical columns to plot
columns = numerical_columns

# Create a subplot figure with 8 rows and 1 column for each distribution
fig = sp.make_subplots(rows=len(columns), cols=1, shared_xaxes=False, subplot_titles=[f'Distribution of {col}' for col in columns])

# Loop through each column and add the histogram and box plot as traces
for i, col in enumerate(columns):
    # Create histogram with KDE
    fig.add_trace(
        go.Histogram(
            x=df1[col],
            # nbinsx=10,
            histnorm='density',
            marker_color="#ffa300",
             marker_line=dict(color='black', width=1),  # Add border color and width
            opacity=0.7,
            name=f'{col} Histogram'
        ),
        row=i+1,
        col=1
    )


# Update layout with titles and spacing
fig.update_layout(
    title_text="Distributions of Numerical Variables",
    height=2000,  # Adjust height to fit all subplots comfortably
    width=800,    # Adjust width as needed
    showlegend=False,
    hovermode='x',  # Enable hover interaction
    margin=dict(t=100, b=20),
)

fig.show()
```

```python

# List of columns to plot
col_bar_plot = ["Garage_Size", "Num_Bathrooms", "Neighborhood_Quality", "Num_Bedrooms"]

# Create a subplot figure with 2 rows and 2 columns
fig = sp.make_subplots(rows=2, cols=2, subplot_titles=col_bar_plot)

# Add each bar plot to the subplot grid
for i, col in enumerate(col_bar_plot):
    # Value counts for each column
    counts = df1[col].value_counts().sort_index()
    
    # Determine row and column index in subplot
    row = i // 2 + 1
    col_idx = i % 2 + 1
    
    # Add bar trace to the figure
    fig.add_trace(
        go.Bar(
            x=counts.index, 
            y=counts.values,
            marker_color="#ffa300",
            name=col
        ),
        row=row,
        col=col_idx
    )

# Update layout settings
fig.update_layout(
    title_text='Bar Plots of Numerical Variables',
    margin=dict(l=20, r=20, b=20, t=70),
    hovermode='x',  # Enable hover interaction
    showlegend=False,
    
)

# Style the axes for each subplot
fig.update_xaxes(title_text="Category", showgrid=True, gridwidth=0.5, gridcolor="gray")
fig.update_yaxes(title_text="Count", showgrid=True, gridwidth=0.5, gridcolor="gray")

fig.show()
```

```python

# Define columns and target variable
columns = numerical_columns
target_var = 'House_Price'

# Create a 4x2 grid for subplots
fig = sp.make_subplots(rows=4, cols=2, subplot_titles=[f'Correlation between {col} and {target_var}' for col in columns])

# Loop through each column to create scatter plots with correlation annotations
for i, col in enumerate(columns):
    # Calculate correlation
    correlation = df1[target_var].corr(df1[col])
    corr_text = f'r = {correlation:.2f}'
    
    # Row and column index for the subplot
    row = i // 2 + 1
    col_idx = i % 2 + 1

    # Add scatter trace
    fig.add_trace(
        go.Scatter(
            x=df1[target_var],
            y=df1[col],
            mode='markers',
            marker=dict(color='#ffa300'),
            name=f'{col} vs {target_var}',
        ),
        row=row,
        col=col_idx
    )
    
    # Add correlation annotation to each subplot
    fig.add_annotation(
        x=df1[target_var].min(),
        y=df1[col].max() * 0.9,
        text=corr_text,
        showarrow=False,
        font=dict(size=12, color='black'),
        align='center',
        bgcolor='#d9d9d9',
        opacity=0.8,
        row=row,
        col=col_idx
    )

# Update layout
fig.update_layout(
    title_text="Correlation of Numerical Variables with Total",
    height=1200,
    width=1200,
    showlegend=False,
)

# Update axes titles
for i, col in enumerate(columns):
    row = i // 2 + 1
    col_idx = i % 2 + 1
    fig.update_xaxes(title_text=target_var, row=row, col=col_idx)
    fig.update_yaxes(title_text=col, row=row, col=col_idx)

fig.show()
```
