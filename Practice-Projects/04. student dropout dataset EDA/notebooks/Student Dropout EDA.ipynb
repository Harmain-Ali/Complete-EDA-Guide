```python
print(f"IT IS A CATAGORICAL VARIABLE. \nUNIQUE VALUES IN {df_coloumns[0]} :{df[df_coloumns[0]].unique()}")
```

```python
# Create bar chart figure
col_values = df[df_coloumns[0]].value_counts()
col_values = (col_values*100)//df.shape[0]
col_groups = col_values.index
fig = go.Figure(go.Bar(
    x= col_groups,
    y=col_values,
    marker_color="brown"
))

# Set overall title and axes labels
fig.update_layout(
    title_text=f'Bar Chart of {df_coloumns[0]}',
    xaxis_title='school name',
    yaxis_title='percentage of students',
)

# Style subplot
fig.update_layout(
    hovermode='x',  # Enable hover interaction
)

fig.show()
```

```python

school_col = df_coloumns[0]  # School column
gender_col = df_coloumns[1]  # Gender column

# Group by school and gender, and count the percentage of occurrences
gender_counts = df.groupby([school_col, gender_col]).size().fillna(0)
gender_counts = ((gender_counts*100)//df.shape[0]).unstack()

# Create the figure
fig = go.Figure()

# Add bars for each gender as separate traces, creating a stacked effect
for gender in gender_counts.columns:
    fig.add_trace(
        go.Bar(
            x=gender_counts.index,
            y=gender_counts[gender],
            name=str(gender),
            marker_color="orange" if gender == "M" else "deeppink"  # Different colors for each gender
        )
    )

# Update layout with stacked option
fig.update_layout(
    title_text=f'Bar Chart of {school_col} by {gender_col}',
    xaxis_title='School Name',
    yaxis_title='percentage of gender of Students',
    hovermode='x',
)

fig.show()
```

```python

# Define columns and target variable
columns = numerical_columns
target_var = 'House_Price'

# Create a 4x2 grid for subplots
fig = sp.make_subplots(rows=4, cols=2, subplot_titles=[f'Correlation between {col} and {target_var}' for col in columns])

# Loop through each column to create scatter plots with correlation annotations
for i, col in enumerate(columns):
    # Calculate correlation
    correlation = df1[target_var].corr(df1[col])
    corr_text = f'r = {correlation:.2f}'
    
    # Row and column index for the subplot
    row = i // 2 + 1
    col_idx = i % 2 + 1

    # Add scatter trace
    fig.add_trace(
        go.Scatter(
            x=df1[target_var],
            y=df1[col],
            mode='markers',
            marker=dict(color='#ffa300'),
            name=f'{col} vs {target_var}',
        ),
        row=row,
        col=col_idx
    )
    
    # Add correlation annotation to each subplot
    fig.add_annotation(
        x=df1[target_var].min(),
        y=df1[col].max() * 0.9,
        text=corr_text,
        showarrow=False,
        font=dict(size=12, color='black'),
        align='center',
        bgcolor='#d9d9d9',
        opacity=0.8,
        row=row,
        col=col_idx
    )

# Update layout
fig.update_layout(
    title_text="Correlation of Numerical Variables with Total",
    height=1200,
    width=1200,
    showlegend=False,
)

# Update axes titles
for i, col in enumerate(columns):
    row = i // 2 + 1
    col_idx = i % 2 + 1
    fig.update_xaxes(title_text=target_var, row=row, col=col_idx)
    fig.update_yaxes(title_text=col, row=row, col=col_idx)

fig.show()
```
