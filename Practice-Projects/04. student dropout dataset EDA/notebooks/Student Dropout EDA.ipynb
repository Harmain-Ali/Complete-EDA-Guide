```python
print(f"IT IS A CATAGORICAL VARIABLE. \nUNIQUE VALUES IN {df_coloumns[0]} :{df[df_coloumns[0]].unique()}")
```

```python
# Create bar chart figure
col_values = df[df_coloumns[0]].value_counts()
col_values = (col_values*100)//df.shape[0]
col_groups = col_values.index
fig = go.Figure(go.Bar(
    x= col_groups,
    y=col_values,
    marker_color="brown"
))

# Set overall title and axes labels
fig.update_layout(
    title_text=f'Bar Chart of {df_coloumns[0]}',
    xaxis_title='school name',
    yaxis_title='percentage of students',
)

# Style subplot
fig.update_layout(
    hovermode='x',  # Enable hover interaction
)

fig.show()
```

```python

school_col = df_coloumns[0]  # School column
gender_col = df_coloumns[1]  # Gender column

# Group by school and gender, and count the percentage of occurrences
gender_counts = df.groupby([school_col, gender_col]).size().fillna(0)
gender_counts = ((gender_counts*100)//df.shape[0]).unstack()

# Create the figure
fig = go.Figure()

# Add bars for each gender as separate traces, creating a stacked effect
for gender in gender_counts.columns:
    fig.add_trace(
        go.Bar(
            x=gender_counts.index,
            y=gender_counts[gender],
            name=str(gender),
            marker_color="orange" if gender == "M" else "deeppink"  # Different colors for each gender
        )
    )

# Update layout with stacked option
fig.update_layout(
    title_text=f'Bar Chart of {school_col} by {gender_col}',
    xaxis_title='School Name',
    yaxis_title='percentage of gender of Students',
    hovermode='x',
)

fig.show()
```
