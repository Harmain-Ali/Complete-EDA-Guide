<b>1. LINE CHART</b></br>

<b>Definition: </b>A line chart, also known as a line plot or line graph, is a graphical representation that uses points connected by line segments to show changes in values over a continuous progression, often represented by the x-axis, such as time. The y-axis represents the values for a metric of interest across that progression.</br>

<b>Variable usage: </b> Numerical variable.</br>

<b>Uses: </b> visualizing trends, patterns, Forecasting, Correlation and relationships over time or other continuous variables.</br>
</div>

```python
# LINE CHART SYNTAX IN MATPLOTLIB

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size, )
plt.suptitle('LINE CHART')

columns = ['Research', 'Talent']

for col, ax in zip(columns, axes.flatten()):
    ax.plot(df['Total score'], df[col])
    ax.set_title(f'Line chart of {col} and Total score')
    ax.set_xlabel('Total score')
    ax.set_ylabel(col)


plt.show()
```

```python
# LINE CHART SYNTAX IN SEABORN

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size)
plt.suptitle('LINE CHART')

columns = ['Research', 'Talent']

for col, ax in zip(columns, axes.flatten()):
    sns.lineplot(data=df, x='Total score', y=col, ax=ax)
    ax.set_title(f'Line chart of {col} and Total score')

plt.show()
```

```python
# LINE PLOT SYNTAX IN PLOTLY

columns = ['Research', 'Talent']

# Define layout with two subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=[f'{col} and Total Score' for col in columns])  # Use a list comprehension


# Add traces for each column
for i, col in enumerate(columns):
    fig.add_trace(
        go.Scatter(
            x=df["Total score"],
            y=df[col],
            mode="lines",
            name=col,
            marker_symbol="circle",
            marker_size=10,
            line_color= '#ffa300'
        ),
        row=1,
        col=i+1
    )

# Set overall title and axes labels
fig.update_layout(
    title_text="LINE CHART",
    showlegend=False
)

# Style subplots
fig.update_layout(
    hovermode='x',  # Enable hover interaction
    margin=dict(l=20, r=20, b=20, t=50)
)

# Set minor grid for both subplots
fig.update_xaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")

fig.show()
```

<b>2. SCATTER PLOT</b></br>

<b>Definition: </b>A scatter plot is a chart type used to observe and visually display the relationship between two variables. It represents data points as dots on a two-dimensional plane, with each dot indicating the values of an individual data point. The horizontal axis represents one variable, while the vertical axis represents the other variable. Scatter plots are also known as scattergrams, scatter graphs, or scatter charts.</br>

<b>Variable usage: </b> Numerical variable.</br>

<b>Uses: </b>  Scatter plots are primarily used in Relationship Analysis, Pattern Identification, Outlier Detection, Correlation Assessment and Model Validationsis.</br>
</div>

```python
# SCATTER PLOT SYNTAX IN MATPLOTLIB

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size, )
plt.suptitle('SCATTER PLOT')

columns = ['Research', 'Talent']

for col, ax in zip(columns, axes.flatten()):
    ax.scatter(df['Total score'], df[col])
    ax.set_title(f'Scatter plot of {col} and Total score')
    ax.set_xlabel('Total score')
    ax.set_ylabel(col)

plt.show()
```

```python
# LINE CHART SYNTAX IN SEABORN

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size)
plt.suptitle('SCATTER PLOT')

columns = ['Research', 'Talent']

for col, ax in zip(columns, axes.flatten()):
    sns.scatterplot(data=df, x='Total score', y=col, ax=ax)
    ax.set_title(f'Scatter plot of {col} and Total score')
    ax.set_xlabel('Total score')
    ax.set_ylabel(col)


plt.show()
```

```python
# SCATTER PLOT SYNTAX IN PLOTLY

columns = ['Research', 'Talent']

# Create figure with subplots and titles
fig = make_subplots(rows=1, cols=2, subplot_titles=[f'Scatter plot of {col} and Total score' for col in columns])

# Add scatter traces for each column
for i, col in enumerate(columns):
    fig.add_trace(
        go.Scatter(
            x=df['Total score'],
            y=df[col],
            mode='markers',
            marker=dict(
                size=10,
                color= '#ffa300',
                line=dict(color= '#4a4a4a', width=1)
            ),
            name=col
        ),
        row=1, col=i+1
    )

# Set overall title and axes labels
fig.update_layout(
    title_text='SCATTER PLOT',
    xaxis_title='Total score',
    yaxis_title='Score',
    showlegend=False
)

# Style subplots
fig.update_layout(
    hovermode='x',  # Enable hover interaction
    margin=dict(l=20, r=20, b=20, t=70)
)

# Set minor grid for both subplots
fig.update_xaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")

fig.show()
```

<b>3. AREA PLOT</b></br>

<b>Definition: </b>An area plot, also known as a stacked area plot or an area chart, is a type of graph that displays the cumulative values of different variables over a continuous interval or time period. It is similar to a line graph, but with the area between the line and the x-axis filled with color. Each variable is represented by a colored area, and the total area represents the cumulative value of all the variables at each point.</br>

<b>Variable usage: </b> Numerical variable.</br>

<b>Uses: </b>  Area plots are useful for Comparison of multiple variables, Tracking changes over time and Highlighting proportions.</br>
</div>

```python
# AREA PLOT SYNTAX IN MATPLOTLIB

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size)
plt.suptitle('AREA PLOT')

columns = ['Research', 'Talent']

for col, ax in zip(columns, axes.flatten()):
    ax.plot(df['Total score'],df[col])
    ax.fill_between(df['Total score'], df[col], alpha=.4)
    ax.set_title(f'Area plot of {col} and Total score')
    ax.set_xlabel('Total score')
    ax.set_ylabel(col)

plt.show()
```

```python
# AREA PLOT SYNTAX IN SEABORN

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size)
plt.suptitle('AREA PLOT')

columns = ['Research', 'Talent']

for col, ax in zip(columns, axes.flatten()):
    sns.lineplot(data=df, x='Total score', y=col, ax=ax)
    ax.fill_between(df['Total score'], df[col], alpha=0.4)
    ax.set_title(f'Area plot of {col} and Total score')
    ax.set_xlabel('Total score')
    ax.set_ylabel(col)


plt.show()
```

```python
# AREA PLOT SYNTAX IN PLOTLY

columns = ['Research', 'Talent']

# Create figure with subplots and titles
fig = make_subplots(rows=1, cols=2, subplot_titles=[f'Area plot of {col} and Total score' for col in columns])

# Add area traces for each column
for i, col in enumerate(columns):
    fig.add_trace(
        go.Scatter(
            x=df['Total score'],
            y=df[col],
            mode='lines',
            fill='tozeroy',
            line=dict(color="#ffa300"),
            name=col,
        ),
        row=1, col=i+1
    )

# Set overall title and axes labels
fig.update_layout(
    title_text='AREA PLOT',
    showlegend=False
)

# Style subplots
fig.update_layout(
    hovermode='x',  # Enable hover interaction
    margin=dict(l=20, r=20, b=20, t=70)
)

# Set minor grid for both subplots
fig.update_xaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")

fig.show()
```

<b>3. Histogram</b></br>

<b>Definition: </b>A histogram is a graphical representation that displays the distribution of continuous or discrete data. It consists of a series of bars, where the height of each bar represents the frequency or count of data points falling within a specific range or bin.</br>

<b>Variable usage: </b> Numerical variable.</br>

<b>Uses: </b>  Histograms are commonly used in Distribution Analysis and Identifying Outliers.</br>
</div>

```python
# HISTOGRAM SYNTAX IN MATPLOTLIB
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size)
plt.suptitle('Histogram')

columns = ['Research','Talent']

for col, ax in zip(columns, axes.flatten()):
    ax.hist(df[col], bins=20, density=True)
    ax.set_title(f'Distribution of {col}')

plt.show()
```

```python
# HISTOGRAM SYNTAX IN SEABORN
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size)
plt.suptitle('Historam')

columns = ['Research', 'Talent']

for col, ax in zip(columns, axes.flatten()):
    sns.histplot(data=df, x=col, bins=20, kde=True, ax=ax)
    ax.set_title(f'Distribution of {col}')


plt.show()
```

```python
# HISTOGRAM SYNTAX IN PLOTLY

columns = ['Research', 'Talent']

# Create figure with subplots and titles
fig = make_subplots(rows=1, cols=2, subplot_titles=[f'Distribution of {col}' for col in columns])

# Add histogram and KDE traces for each column
for i, col in enumerate(columns):
    fig.add_trace(
        go.Histogram(
            x=df[col],
            nbinsx=20,  # Specify number of bins
            marker_color="#ffa300",
            marker_line_color= '#4a4a4a',
            opacity=0.8,
            name='Histogram'
        ),
        row=1, col=i+1
    )

# Set overall title and axes labels
fig.update_layout(
    title_text='Histogram',
    showlegend=False
)

# Style subplots
fig.update_layout(
    hovermode='x',  # Enable hover interaction
    margin=dict(l=20, r=20, b=20, t=70)
)

# Set minor grid for both subplots
fig.update_xaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")

fig.show()
```

<b>2. BAR CHART</b></br>

<b>Definition: </b>A bar chart, also known as a bar graph, is a graphical representation that uses rectangular bars of varying lengths to display and compare categorical data. The length of each bar corresponds to the value or frequency of a specific category. Bar charts can be created with vertical bars, horizontal bars, grouped bars, or stacked bars, depending on the type of data being represented.</br>

<b>Variable usage: </b> Categorical variable + Frequency.</br>

<b>Uses: </b>  Bar charts are commonly used in Comparison, Trends, Ranking and Comparison of Subgroups of catagorical data.</br>
</div>

```python
categories = df.value_counts('Region').index
counts = df['Region'].value_counts()
```

```python
# BAR CHART SYNTAX IN MATPLOTLIB

fig, ax = plt.subplots(figsize=fig_size)
plt.suptitle('Bar chart')

ax.bar(categories,counts)
plt.show()
```

```python
# BAR CHART SYNTAX IN SEABORN

fig, ax = plt.subplots(figsize=fig_size)
plt.suptitle('Bar chart')
sns.barplot(x=categories, y=counts, ax=ax)
plt.show()
```

```python
# BAR CHART SYNTAX IN PLOTLY

# Create bar chart figure
fig = go.Figure(go.Bar(
    x=categories,
    y=counts,
    marker_color="#ffa300"
))

# Set overall title and axes labels
fig.update_layout(
    title_text='Bar Chart',
    xaxis_title='Category',
    yaxis_title='Count',
    margin=dict(l=20, r=20, b=20, t=70)
)

# Style subplot
fig.update_layout(
    hovermode='x',  # Enable hover interaction
    xaxis=dict(showgrid=True, gridwidth=0.5, gridcolor="gray"),  # Add minor grid lines
    yaxis=dict(showgrid=True, gridwidth=0.5, gridcolor="gray")
)

fig.show()
```

<b>4. BOX PLOT</b></br>

<b>Definition: </b>A box plot, also known as a box-and-whisker plot, is a graphical representation that displays the distribution of a continuous variable. It provides a summary of the data's central tendency, spread, and any potential outliers.</br>

<b>Variable usage: </b> Numerical and Categorical variable.</br>

<b>Uses: </b>  Box plots are commonly used in Distribution Analysis, Data Summary and Identifying Outliers.</br>
</div>

```python
# BOX PLOT SYNTAX IN MATPLOTLIB

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size)
plt.suptitle('Box plot')

columns = ['Research', 'Talent']
for col, ax in zip(columns, axes.flatten()):
    ax.grid(which= 'major')
    ax.boxplot(df[col],vert = False, showmeans= True)
    ax.set_title(f'Distribution of {col}')

plt.show()
```

```python
# BOX PLOT SYNTAX IN SEABORN

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size)
plt.suptitle('Box plot')

columns = ['Research', 'Talent']
for col, ax in zip(columns, axes.flatten()):
    sns.boxplot(x=df[col], ax=ax, orient='h', showmeans=True)
    ax.grid(which= 'major')
    ax.set_title(f'Distribution of {col}')

plt.show()
```

```python
# BOX PLOT SYNTAX IN PLOTLY

columns = ['Research', 'Talent']

# Create figure with subplots and titles
fig = make_subplots(rows=1, cols=2, subplot_titles=[f'Distribution of {col}' for col in columns])

# Add box plots with mean markers for each column
for i, col in enumerate(columns):
    fig.add_trace(
        go.Box(
            y=df[col],  # Note: y-axis for horizontal box plots
            boxpoints='all',  # Show all points
            marker_color="#ffa300",
            showlegend=False
        ),
        row=1, col=i+1
    )
# Set overall title and axes labels
fig.update_layout(
    title_text='Box Plot',
    margin=dict(l=20, r=20, b=20, t=70)
)

fig.update_xaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")

fig.show()
```

<b>6. VIOLIN PLOT</b></br>

<b>Definition: </b>A violin plot is a type of chart used to visualize the distribution of numerical data. It combines aspects of a box plot and a kernel density plot to provide a comprehensive view of the data's summary statistics and density.</br>

<b>Variable usage: </b> Numerical and Categorical variable.</br>

<b>Uses: </b>   Violin plots are useful for Distribution Comparison, Summary Statistics, Outlier Detection and Density Estimation of categorical variable.</br>
</div>

```python
# VIOLIN PLOT SYNTAX IN MATPLOTLIB

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size)
plt.suptitle('Violin plot')

columns = ['Research', 'Talent']
for col, ax in zip(columns, axes.flatten()):
    parts = ax.violinplot(df[col], vert=False, showmeans=False, showmedians=True)
    ax.set_title(f'Distribution of {col}')

plt.show()
```

```python
# VIOLIN PLOT SYNTAX IN SEABORN

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=fig_size)
plt.suptitle('Violin plot')

columns = ['Research', 'Talent']
for col, ax in zip(columns, axes.flatten()):
    sns.violinplot(x=df[col], ax=ax, inner="box",fill=False)
    ax.minorticks_on()
    ax.grid(which='major', linestyle='solid', linewidth='0.5', color='gray')
    ax.set_title(f'Distribution of {col}')

plt.show()
```

```python
# VIOLIN PLOT SYNTAX IN PLOTLY

columns = ['Research', 'Talent']

# Create figure with subplots and titles
fig = make_subplots(rows=1, cols=2, subplot_titles=[f'Distribution of {col}' for col in columns])

# Add violin traces for each column
for i, col in enumerate(columns):
    fig.add_trace(
        go.Violin(
            y=df[col],
            box_visible=True,  # Show box outlines
            line_color="#ffa300",
            meanline_visible=True,  # Show mean line
            meanline_color='black',
            showlegend=False
        ),
        row=1, col=i+1
    )

# Set overall title and axes labels
fig.update_layout(
    title_text='Violin Plot',
    margin=dict(l=20, r=20, b=20, t=70)
)
fig.update_xaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor="gray")

fig.show()
```

<b>7. HEATMAP</b></br>

<b>Definition: </b>A heatmap is a graphical representation of data where individual values are represented as colors. It is primarily used to show the magnitude of a phenomenon as color in two dimensions, and the variations in color represent the data magnitude. Heatmaps are great for visualizing matrix-like data, where both the x and y-axes represent categorical variables, and the color intensity shows the relationship between them.</br>

<b>Variable usage: </b> Categorical and Numerical variables.</br>

<b>Uses: </b>   Heatmaps are particularly useful for:
Visualizing Correlation Matrices,
Showing Frequency Distributions,
Exploring Data Relationships,
Identifying Patterns and Anomalies,
Presenting Categorical Data with Color Intensity.</br>
</div>

```python
# HEATMAP SYNTAX IN MATPLOTLIB
coloumns = ['Total score','Talent','Infrastructure','Operating Environment','Research','Development','Government Strategy','Commercial']
correlation_matrix = df[coloumns].corr()
fig, ax = plt.subplots(figsize=(10, 8))
plt.suptitle('HEATMAP')

# Assume 'data' is a correlation matrix or a 2D matrix
heatmap = ax.imshow(correlation_matrix , cmap='coolwarm', interpolation='nearest')

# Add color bar for reference
cbar = plt.colorbar(heatmap)
# # Add labels to axes (if needed)
ax.set_xticks(range(len(correlation_matrix.columns)))
ax.set_yticks(range(len(correlation_matrix.columns)))
ax.set_xticklabels(correlation_matrix.columns,rotation=45)
ax.set_yticklabels(correlation_matrix.columns)
plt.show()
```

```python
# HEATMAP SYNTAX IN SEABORN

# columns = ['Total score', 'Talent', 'Infrastructure', 'Operating Environment', 'Research', 'Development', 'Government Strategy', 'Commercial']
# correlation_matrix = df[columns].corr()

# Set the figure size
plt.figure(figsize=(10, 8))
plt.suptitle('HEATMAP')

# Plot the heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

# Rotate x-ticks for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# HEATMAP SYNTAX IN plotly

columns = ['Total score', 'Talent', 'Infrastructure', 'Operating Environment', 'Research', 'Development', 'Government Strategy', 'Commercial']
correlation_matrix = df[columns].corr()

# Create the heatmap
fig = go.Figure(data=go.Heatmap(
    z=correlation_matrix.values,
    x=correlation_matrix.columns,
    y=correlation_matrix.columns,
    colorscale='ylorrd',
    colorbar=dict(title='Correlation'),
    zmin=-1,  # Set the min value of the color scale
    zmax=1,   # Set the max value of the color scale
))

# Update layout for title and x/y axis labels
fig.update_layout(
    title='HEATMAP',
    xaxis_title='Features',
    yaxis_title='Features',
    xaxis=dict(tickangle=45)  # Rotate x-ticks
)

# Show the plot
fig.show()
```
