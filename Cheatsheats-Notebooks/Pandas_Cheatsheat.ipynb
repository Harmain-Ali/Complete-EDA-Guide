<p>
The Dataset "AI Global index" includes The Global AI Index itself and seven indicators affecting the Index on 62 countries, as well as general information about the countries (region, cluster, income group and political regime).

The Global AI Index is the first index to benchmark nations on their level of investment, innovation and implementation of artificial intelligence.

Talent, Infrastructure and Operating Environment are the factors of AI Implementation group of indicators, which represents the application of artificial intelligence by professionals in various sectors, such as businesses, governments, and communities.
The data as described as:
<ul>
    <li>Talent indicator focuses on the availability of skilled practitioners for the provision of artificial intelligence solutions.</li>
    <li>Infrastructure indicator focuses on the reliability and scale of access infrastructure, from electricity and internet, to super computing capabilities.</li>
    <li>Operating Environment indicator focuses on the regulatory context, and public opinion surrounding artificial intelligence.</li>
    <li>Research indicator focuses on the extent of specialist research and researchers; investigating the amount of publications and citations in credible academic journals.</li>
    <li>Development indicator focuses on the development of fundamental platforms and algorithms upon which innovative artificial intelligence projects rely.</li>
    <li>Government Strategy indicator focuses on the depth of commitment from national government to artificial intelligence; investigating spending commitments and national strategies.</li>
    <li>Commercial indicator focuses on the level of startup activity, investment and business initiatives based on artificial intelligence.</li>
    
</ul>

**7. Cleaning The Data**

`df.isnull()` : Checks for null Values, Returns Boolean Array.

`df.notnull()` : Opposite of is null()

`df.dropna()` : Drops all rows that contain null values

`df.fillna(x)` : Replaces all null values with x that can be a value or any statistical function.

`df.ffill()` : Replace null values with previous non null values.

`df.bfill()` : it will replace null values with next non null values.

`df.drop()` : remove rows or columns based on specified labels.

`df.drop_duplicates()` :  Delete the rows which are identical.

`df.sort_values()` : Sort the DataFrame by the specified column(s) or index. It allows for sorting in ascending or descending order.

`df.reset_index()` : Reset the index of a DataFrame.

`df.set_index()` : Allows you to designate a specific column as the index of the DataFrame.

`df.rename()` : Rename columns and indexes in a DataFrame

`df.replace()` :  Replace values within a DataFrame. 


```python
# The df.isnull() method in Pandas is used to identify missing values within a DataFrame.
# When this method is called, it returns a DataFrame of the same shape as the original,
# where each cell is either True or False, indicating whether the corresponding cell in the original DataFrame is null or not.
df.isnull()
```

```python
# The df.notnull() method in Pandas is used to identify non-null values within a DataFrame.
# When this method is called, it returns a DataFrame of the same shape as the original, 
# where each cell is either True or False, indicating whether the corresponding cell in the original DataFrame is not null.
df.notnull()
```

```python
# The df.dropna() method in Pandas is used to remove rows or columns with missing values (NaN, NaT, None, etc.) from a DataFrame. 
# It provides flexibility in specifying the axis, how to handle the removal, and the subset of columns to consider.
df.dropna()
```

```python
# The df.fillna(x) method in Pandas is used to replace NaN (missing) values in a DataFrame with a specified value,
# providing flexibility in the form of a scalar, a dictionary, or a DataFrame.
df.fillna(55) 
```

```python
# Filling null values with mean, median, mode
mean = df["Talent"].mean() # we can use median(), mode()[0], std() and much more methods
del_df = df["Talent"].fillna(mean)
```

```python
# The df.ffill() method in Pandas is used to forward-fill missing values in a DataFrame. 
# When this method is called, it propagates the last valid observation forward to fill the missing values.
df.ffill()
```

```python
# The df.bfill() method in Pandas is used to backward-fill missing values in a DataFrame.
# When this method is applied, it replaces the NULL values with the values from the next row 
# (or next column, if the axis parameter is set to 'columns').
df.bfill() 
```

```python
# The df.drop() method in Pandas is used to remove rows or columns from a DataFrame based on specified labels. 
# It provides flexibility in specifying the axis (rows or columns) and the labels to drop.

#df1 = df1.drop("Research",axis=1)
df1.head(2)
```

```python
# The df.drop_duplicates() method in Pandas is used to remove duplicate rows from a DataFrame based on the values in the specified columns.
# By default, it removes duplicate rows based on all columns,but it also provides the flexibility to specify a subset of columns
# to consider for identifying duplicates.
df.drop_duplicates() 
```

```python
# The df.sort_values() method in Pandas is used to sort the DataFrame by the specified column(s) or index.
# It allows for sorting in ascending or descending order and provides various parameters for customization.
df = df.sort_values("Total score", ascending= True)
```

```python
# The df.reset_index() method in Pandas is used to reset the index of a DataFrame, which can be particularly useful after data manipulation or
# filtering operations that may have altered the original index. This method returns a new DataFrame with a new numeric index,
# starting from 0, and the original index becomes a new column in the DataFrame, labeled with the name 'index'.
df.reset_index()
```

```python
# To set the index of a DataFrame in Pandas, you can use the set_index() method.
# This allows you to designate a specific column as the index of the DataFrame.

# df.set_index("Total score")
```

```python
# The rename() method in Pandas is a powerful tool for renaming columns and indexes in a DataFrame.
# It allows for flexible and efficient renaming of specific columns or indexes,
# as well as the ability to rename all columns or indexes at once.

# To rename columns in a Pandas DataFrame, you can use the rename() method with a dictionary 
# that maps the old column names to the new column names.
df.rename(columns={"Talent" :"Skills"}, inplace=True)
df.head(2)
```

```python
# The replace() method in Pandas is a versatile tool for replacing values within a DataFrame. 
# It allows for simple value replacement as well as more complex replacements based on specific conditions.
df["Region"].replace(["Asia-Pacific","Middle East"], ["Asia","Mid-East"], inplace=True)
df.head()
```
